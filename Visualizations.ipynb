## Visualizations for gender classification on twitter

For all visualizations we've used counters for male and female. As in the main code, we have divided the number of found features by the total amount of tweets. These features are then used for the graphs. To find these features we have used functions from the main code. 

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("whitegrid")
import numpy as np
```

#### 1. Exclamation marks in tweets 
##### 1.1. Tweets either with or without exclamation mark

```python
#Stacked bar graph showing tweets with and without exclamation marks

N = 2
tweets_without_excl = (3826, 4019)
tweets_with_excl = (509, 671)
fig1 = plt.figure()
ind = np.arange(N)
width = 0.5

p1 = plt.bar(ind,tweets_without_excl, width, color='#2276D2')
p2 = plt.bar(ind, tweets_with_excl, width, bottom = tweets_without_excl, color='#F0A01D')

plt.ylabel('Number of tweets')
plt.title('Number of tweets without and with exclamation marks')
plt.xticks(ind,('Male', 'Female'))
plt.yticks(np.arange(0, 6500, 500))
plt.legend((p1[0], p2[0]), ('Tweets without exclamation mark', 'Tweets with exclamation mark'))
plt.show()
```

This bar graph shows that the female twitter-users have more tweets in general, but also have a much larger amount of tweets containing an exclamation mark than the male twitter-users have. 

##### 1.2 Different amount of exclamation marks used 

```python
#Bar graph showing tweets with different number of exclamation marks M/F

n_groups = 6
tweets_male = (7.7, 2.5, 1.0, 0.2, 0.1, 0.1)
tweets_female = (8.9, 3.3, 0.9, 0.7, 0.3, 0.1)
fig2 = plt.figure()

fig2, ax = plt.subplots()
index = np.arange(n_groups)
bar_width = 0.35
opacity = 0.9

rects1 = plt.bar(index, tweets_male, bar_width, alpha=opacity, color='#2276D2', label='Male')
rects2 = plt.bar(index + bar_width, tweets_female, bar_width, alpha=opacity, color='#F0A01D', label='Female')

plt.xlabel('Number of exclamation marks')
plt.ylabel('Number of tweets (%)')
plt.title('Percentage of tweets with number of exclamation marks')
plt.xticks(index +bar_width, ('1!', '2!', '3!', '4!', '5!', '6!'))
plt.legend()

plt.tight_layout()
plt.show()
```

From this bar graph we can see that in general, females tweet a tweet containing one or more exclamation marks more often than males. Interestingly, males had more tweets that contained three exclamation marks, but with all other amounts of exclamation marks, females produce more tweets

#### 4. Use of first person expressions


We wanted to look at the difference in the use of first person expressions. To do this we needed to look at the different ways to spell for example I (I/i). We looked at all expressions in isolation (so placing a space before and after the expression within the string) and using capitals at the beginning of the expression. The ways we spelled it can be seen below: 

" I ", "I ", " i "

" I'm ", "I'm ", " im ", " Im ", " i'm "

"Me", " Me ", " me "

" my ", " My ", "My "

"I have ", " I have ", " i have "

"I am ", " I am ", " i am "

We left out certain ways of spelling to avoid including an expression such as "mystery" which would be included if we have also spelled 'my' as: "my" or " my". 

```python
n_groups = 6
tweets_male = (37.8, 8.8, 5.6, 8.9, 2.0, 1.2)
tweets_female = (48.9, 13.3, 9.9, 15.0, 3.2, 1.8)
fig5 = plt.figure()

fig5, ax = plt.subplots()
index = np.arange(n_groups)
bar_width = 0.35
opacity = 0.9

rects1 = plt.bar(index, tweets_male, bar_width, alpha=opacity, color='#2276D2', label='Male')
rects2 = plt.bar(index + bar_width, tweets_female, bar_width, alpha=opacity, color='#F0A01D', label='Female')

plt.xlabel('First person expressions')
plt.ylabel('Number of tweets (%)')
plt.title('Male/female comparison of the use of first person expressions')
plt.xticks(index +bar_width, ('I', "I'm", 'me', 'my', 'I have', 'I am'))
plt.legend()

plt.tight_layout()
plt.show()
```

From this graph we can see that on all the expressions we included, female twitter users use them overwhelmingly more often than men.
